## **Models**

**Libraries Used**

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

**Split data into training and test sets**

**I chose to use the variables** ('SEX', 'DESIGNATION', 'UNIT', 'RATINGS', 'TOTAL_EXP') because they showed a strong impact in the EDA. Although the 'RATINGS' column did not show relevance in the EDA charts, we understand that it may be an important variable when determining a professional's salary.

**I did not use the variables**('DOJ', 'CURRENT DATE', 'FIRST NAME', 'LAST NAME', 'PAST EXP', 'TENURE', 'AGE'):

The variables 'FIRST NAME' and 'LAST NAME' would not represent any significant change, as the goal of the model is to generalize and not to seek a specific data point within the dataframe.
The variables 'DOJ' and 'CURRENT DATE' were replaced by the variable 'TENURE', which was combined with 'PAST EXP' to obtain the variable 'TOTAL_EXP', which is used in the training.
The variable 'AGE' would not be relevant, as although 'AGE' tends to have a correlation with 'SALARY', I do not believe there is a direct causation.

```python
df = df.drop(columns=['DOJ', 'CURRENT DATE','FIRST NAME','LAST NAME','PAST EXP','TENURE','AGE'])
```

```python
X = df.drop(columns=["SALARY"])
y = df["SALARY"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
s1 = ColumnTransformer([
                        ('ord', OrdinalEncoder(categories=[['Analyst', 'Senior Analyst', 'Associate', 'Manager', 'Senior Manager', 'Director']]), ['DESIGNATION']),
                        ('onehot', OneHotEncoder(), ['UNIT', 'SEX'])
                        ], remainder='passthrough')

s3 = ColumnTransformer([
                        ('Scale', StandardScaler(), slice(0,122))
                        ], remainder="passthrough")

# List of models to evaluate
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regression": RandomForestRegressor(),
    "Decision Tree Regressor": DecisionTreeRegressor()
}

for name, model in models.items():
    pipeline = Pipeline([
        ("s1", s1),  # Categorical variable coding
        ("s3", s3),  # Feature scaling
        ("s2", model),  # Model
    ])

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f"{name} R2 Score: {r2}")
    print(f"{name} MSE: {mse}")
    print(f"{name} MAE: {mae}\n")
    # print('\n')
```

```python
# Feature Importance of Each Variable for the Decision Tree Regressor Model
if name == "Decision Tree Regressor":
        importances = pipeline.named_steps["s2"].feature_importances_
        print(f"{name} Feature Importances:")
        for feature_name, importance in zip(X_train.columns, importances):
            print(f"    {feature_name}: {importance}")
```
